```python
# 컬럼명 변경: rename()
cctv_df.rename(columns={"구분": "구 별", "2015년 이전 설치된 CCTV ": "2015년 이전"}, inplace=True)
```

```python
cctv_df["2020년 이전"] = cctv_df["2015년 이전"] + cctv_df["2016년"] + cctv_df["2017년"] + cctv_df["2018년"] + cctv_df["2019년"] + cctv_df["2020년"]
```

```python
cctv_df.drop(columns=["2020년 이전"], inplace=True)
```

```python
# 두번째 방법: sum()
cctv_df["2020년 이전"] = cctv_df[["2015년 이전", "2016년", "2017년", "2018년", "2019년", "2020년"]].sum(axis=1)
cctv_df.head(3)
```

```python
# 필요없는 컬럼 제거: '2015년 이전', '2016년', '2017년', '2018년', '2019년', '2020년'
cctv_df.drop(columns=['2015년 이전', '2016년', '2017년', '2018년', '2019년', '2020년'])
```

```python
cctv_df.drop(columns=cctv_df[cctv_df.columns.difference(['구 별', '2020년 이전', '2021년', '2022년', '2023년', '2024년'])], inplace=True)
```

```python
cctv_df.head(3)
```

```python
# 컬럼 추가
# cctv 총계 = '2021년' + '2022년' + '2023년' + '2024년' + '2020년 이전'

cctv_df["CCTV 총 계"] = cctv_df[['2021년', '2022년', '2023년', '2024년', '2020년 이전']].sum(axis=1)
```

```python
cctv_df.head(3)
```

```python
# 구 별 cctv 가장 적게 설치되어 있는 구 조회
cctv_df[min(cctv_df['CCTV 총 계']) == cctv_df['CCTV 총 계']]
# cctv_df['CCTV 총 계'].min()
```

```python
# 구 별 cctv 가장 많이 설치되어 있는 구 조회
cctv_df[max(cctv_df['CCTV 총 계']) == cctv_df['CCTV 총 계']]
```

#### 서울시 거주 인구 데이터 로드

```python
seoul_df = pd.read_excel("./resources/등록인구.xlsx", skiprows=[0, 1, 3], usecols="B, D, G, J")
seoul_df.head(3)
```

```python
# 컬럼명 변경: 구 별, 총인구, 한국인, 외국인
# seoul_df.rename(columns={"동별(2)": "구 별", "소계.1": "총인구", "소계.2": "한국인", "소계.3": "외국인"})
seoul_df.columns = ["구 별", "총인구", "한국인", "외국인"]
seoul_df
```

```python
# 외국인이 제일 많이 사는 구

seoul_df[max(seoul_df['외국인']) == seoul_df['외국인']]
seoul_df.sort_values("외국인", ascending=False).head(5)
```

```python
# 중구 값 변경
# cctv_df - 중 구, seoul_df - 중구
# cctv_df 변경

# 특정 인덱스 조회
cctv_df.loc[1]
cctv_df.iloc[0]

cctv_df.iloc[1, 0]
cctv_df.loc[2, "구 별"] = "중구"
```

```python
cctv_df
```

```python
# cctv_df와 seoul_df 병합

merge_df = pd.merge(seoul_df, cctv_df, on="구 별")
```

```python
# 인덱스 변경: 구 별
# set_index()

merge_df.set_index("구 별", inplace=True)
```

```python
# CCTV 총 계 / 총인구 * 100 = 1인당 CCTV 비율

merge_df["1인당 CCTV 비율"] = merge_df["CCTV 총 계"] / merge_df["총인구"] * 100
```

```python
merge_df
```

#### 상관계수
- -1 ~ 1 사이의 값으로 나타남
- 0에 가까울 수록 상관 관계는 낮아지며, -1이나 1에 가까울 수록 상관 관계는 높아짐
- 증감의 방향이 +인 경우에는 양의 상관 관계, -인 경우에는 음의 상관 관계를 가짐

```python
# CCTV 총 계와 총인구의 상관관계

merge_df["총인구"].corr(merge_df["CCTV 총 계"], method="spearman")
```

```python
# 외국인과 CCTV 총 계의 상관관계

merge_df["외국인"].corr(merge_df["CCTV 총 계"], method="spearman")
```

```python
# 시각화 - 1인당 CCTV 비율
# barh
plt.figure(figsize=(12, 8))

plt.barh(y=merge_df.index, width=merge_df["1인당 CCTV 비율"])
g = plt.title("구 별 CCTV 비율")
```

```python
merge_df["1인당 CCTV 비율"].sort_values().plot.barh(figsize=(12, 8))
g = plt.title("구 별 CCTV 비율")
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=merge_df, x="1인당 CCTV 비율", y=merge_df.index, order=merge_df["1인당 CCTV 비율"].sort_values(ascending=False).index, palette="rocket")
```

#### 5대 범죄

```python
crime_df = pd.read_excel("./resources/5대범죄.xlsx", skiprows=[0, 1, 2, 4], usecols="B:D")
crime_df.head()
```

```python
# 컬럼명: 구 별, 범죄건, 검거건
crime_df.columns = ["구 별", "범죄건", "검거건"]
crime_df.head()
```

```python
# index 변경: 구 별
crime_df.set_index("구 별", inplace=True)
```

```python
crime_df.head()
```

```python
# 검거율 = 검거건 / 범죄건 * 100

crime_df["검거율"] = crime_df["검거건"] / crime_df["범죄건"] * 100
```

```python
crime_df.head()
```

```python
# merge_df와 crime_df 병합

# merge_df: on => 동일 컬럼
merge_crime_df = pd.merge(merge_df, crime_df, left_index=True, right_index=True)
merge_crime_df.head()
```

```python
# CCTV 총 계와 검거건의 상관관계
# CCTV 총 계와 범죄건의 상관관계
# scatter plot

plt.scatter(x=merge_crime_df["검거건"], y=merge_crime_df["CCTV 총 계"])
plt.scatter(x=merge_crime_df["범죄건"], y=merge_crime_df["CCTV 총 계"])
plt.xlabel("범죄건/검거건")
plt.ylabel("CCTV 총 계")
g = plt.legend(["검거건", "범죄건"])
```

```python
ax1 = merge_crime_df.plot.scatter(x="검거건", y="CCTV 총 계", color="red", s=50)
merge_crime_df.plot.scatter(x="범죄건", y="CCTV 총 계", ax=ax1, color="blue", s=30)
plt.legend(["검거건", "범죄건"])
g = plt.xlabel("범죄건/검거건")
```

```python
sns.scatterplot(data=merge_crime_df, x="검거건", y="CCTV 총 계")
sns.scatterplot(data=merge_crime_df, x="범죄건", y="CCTV 총 계")
```
